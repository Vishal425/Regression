```python
data = pd.read_excel('House Price1.xlsx')
```

```python
data['Front Area'] = data['Front Area'].fillna(data['Front Area'].median())
```

```python
data
```

```python
data.isnull().sum()
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
```

```python
IQR = Q3 - Q1
```

```python
data1 = data[~((data<(Q1-1.5*IQR))| (data>(Q3+1.5*IQR))).any(axis=1)]
```

```python
data1
```

```python
x = data1.iloc[:,1:7].values
y = data1.iloc[:,[-1]].values


pd.DataFrame(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
len(x_train)
```

```python
len(y_test)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
pickle.dump(regressor, open('sales.pkl','wb'))
```
